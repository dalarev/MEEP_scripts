```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import meep as mp
```

```python
## Test: create MEEP geometry, output_epsilon, and try to read this MEEP-generated hdf5

resolution = 64
cell_size = mp.Vector3(1,1,0)

geometry = [mp.Cylinder(0.2, material=mp.Medium(epsilon=2.0))]

sim = mp.Simulation(resolution=resolution,
                    cell_size=cell_size,
                    geometry=geometry,
                   )
sim.plot2D()
sim.run(mp.at_beginning(mp.output_epsilon), until=0)
```

```python
sim = mp.Simulation(resolution=resolution,
                    cell_size=cell_size,
                    epsilon_input_file="eps-000000000.h5",
                   )
sim.plot2D()
plt.show()
```
